```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("student_habits_performance.csv")

# Exploratory Data Analysis (EDA) - Stacked Bar Chart Most Context and Grouping for Nonconditional Data
target_col = [
    { 'count_col' : 'gender', 'group_col' : 'part_time_job' },
    { 'count_col' : 'gender', 'group_col' : 'diet_quality' },
    { 'count_col' : 'gender', 'group_col' : 'extracurricular_participation' },
]

for dt in target_col:
    count_col = dt['count_col']
    group_col = dt['group_col']
    
    # Dataset
    context_group_counts = df.groupby([count_col, group_col]).size().reset_index(name='Count')
    stacked_data = context_group_counts.pivot(index=count_col, columns=group_col, values='Count').fillna(0)
    
    # Ordering
    stacked_data['Total'] = stacked_data.sum(axis=1)
    stacked_data = stacked_data.sort_values(by='Total', ascending=False).head(7)
    categories = stacked_data.columns[:-1] 
    counts = stacked_data[categories].values 
    
    # Plotting
    context = stacked_data.index
    fig, ax = plt.subplots(figsize=(10, 6))
    bottom = np.zeros(len(context))
    
    # Plot each group_col as stack
    for group_col_index, group in enumerate(categories):
        p = ax.bar(context, counts[:, group_col_index], label=group, bottom=bottom)
        bottom += counts[:, group_col_index]
        ax.bar_label(p, label_type='center', fmt='%d')
    
    # Customizing the plot
    ax.set_title(f'Total Student by {count_col} and {group_col}')
    ax.set_xlabel(count_col)
    ax.set_ylabel('Number of Student')
    ax.legend(title=group_col, bbox_to_anchor=(1, 1), loc='upper left')
    ax.tick_params(axis='x', rotation=45)
    plt.tight_layout()
    plt.show()
    print("\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Exploratory Data Analysis (EDA) - Stacked Bar Chart Total Student Per Grade And Study Hour Range
df = pd.read_csv("student_habits_performance.csv")

df['Study Hour Range'] = pd.cut(
    df['study_hours_per_day'], 
    bins=[0, 3, 6, 9], 
    labels=['0-2', '3-5', '6-8'],
    right=False,
    include_lowest=True
)
df['Grade'] = pd.cut(
    df['exam_score'], 
    bins=[0, 60, 70, 80, 90, 101],
    labels=['F', 'D', 'C', 'B', 'A'],
    right=False,
    include_lowest=True
)

context_group_counts = df.groupby(['Grade', 'Study Hour Range'], observed=False).size().reset_index(name='Count')
stacked_data = context_group_counts.pivot(index='Grade', columns='Study Hour Range', values='Count').fillna(0)

# Ordering
stacked_data['Total'] = stacked_data.sum(axis=1)
stacked_data = stacked_data.sort_values(by='Total', ascending=False)
categories = stacked_data.columns[:-1] 
counts = stacked_data[categories].values 

# Plotting
context = stacked_data.index
fig, ax = plt.subplots(figsize=(10, 6))
bottom = np.zeros(len(context))

# Plot each group_col as stack
for group_col_index, group in enumerate(categories):
    p = ax.bar(context, counts[:, group_col_index], label=group, bottom=bottom)
    bottom += counts[:, group_col_index]
    ax.bar_label(p, label_type='center', fmt='%d')

# Bar Chart
ax.set_title(f'Total Student Per Grade And Study Hour Range')
ax.set_xlabel('Grade')
ax.set_ylabel('Total')
ax.legend(title='Study Hour Range', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
