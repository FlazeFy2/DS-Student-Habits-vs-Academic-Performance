```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_habits_performance.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['gender','part_time_job','diet_quality','parental_education_level','internet_quality','extracurricular_participation']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"{dt.replace("_"," ").title()} Comparison")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Pie Chart Comparison Exam Score Result
df = pd.read_csv("student_habits_performance.csv")

# Dataset
def scoring_exam(score):
    if score >= 100:
        return "Perfect"
    elif score >= 85:
        return "Excellent"
    elif score >= 70:
        return "Good"
    elif score >= 50:
        return "Average"
    else:
        return "Poor"

# Apply scoring
df["exam_score_result"] = df["exam_score"].apply(scoring_exam)

top_context = df["exam_score_result"].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Exam Score Result Comparison")
plt.ylabel("")
plt.show()
```
